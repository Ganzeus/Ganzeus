```python
# 自定义训练集
X, y = make_classification(n_samples=300, n_features=2, n_redundant=0, n_informative=2, random_state=22, n_clusters_per_class=1, scale=100)     # 随机数种子手动设置为22
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
X = preprocessing.minmax_scale(X, feature_range=(0, 1))     # 0-1标准化
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

clf = SVC()     # classfier = Support Vector Classfier
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))        # 计算模型准确率
```

## Cross Validation
+ 在数据集中多次随机划分训练集与测试集，每次划分都重新训练模型
+ 最后将每次准确率取平均，就称为交叉验证（cross validation）

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()          # 准备数据集
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=4)   # 划分训练集与测试集
knn = KNeighborsClassifier(n_neighbors=5)       # 定义模型
knn.fit(X_train, y_train)                       # 模型训练
y_pred = knn.predict(X_test)                    # 模型预测
knn.score(X_test, y_test)                       # 输出准确率
```

```python
from sklearn.model_selection import cross_val_score

knn = KNeighborsClassifier(n_neighbors=5)
scores = cross_val_score(knn, X, y, cv=5, scoring="accuracy")   # 自动分成五组，划分五次训练集与测试集，返回5组准确率
scores
```

```python
scores.mean()
```

### 调参-n_neighbors

```python
k_scores = []
for k in range(1, 31):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv = 10, scoring="accuracy")    # for claasification
    # loss = -cross_val_score(knn, X, y, cv=10, scoring='neg_mean_squared_error')  # for regression
    k_scores.append(scores.mean())  

plt.plot(range(1, 31), k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
k_loss = []
for k in range(1, 31):
    knn = KNeighborsClassifier(n_neighbors=k)
    # scores = cross_val_score(knn, X, y, cv = 10, scoring="accuracy")    # for claasification
    loss = -cross_val_score(knn, X, y, cv=10, scoring='neg_mean_squared_error')  # for regression
    k_loss.append(loss.mean())  

plt.plot(range(1, 31), k_loss)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Loss')
plt.show()
```

```python
from sklearn.model_selection import learning_curve
from sklearn.datasets import load_digits    # 数据集（0-9数字）
from sklearn.svm import SVC                 # 模型
import matplotlib.pyplot as plt
import numpy as np

digits = load_digits()  
X, y = digits.data, digits.target
train_sizes, train_loss, test_loss = learning_curve(
    SVC(gamma=0.001), X, y, cv=10, scoring='neg_mean_squared_error', 
    train_sizes=[0.1, 0.25, 0.5, 0.75, 1]
)
train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

plt.plot(train_sizes, train_loss_mean, 'o-', color='r', label='Training')
plt.plot(train_sizes, test_loss_mean, 'o-', color='g', label='Cross-validation')

plt.xlabel("Training examples")
plt.ylabel("Loss")  
plt.legend(loc="best")
plt.show()
```

## 保存模型

```python
from sklearn import svm
from sklearn import datasets

clf = svm.SVC()
iris = datasets.load_iris()
X, y = iris.data, iris.target
clf.fit(X, y)

# 方法一：pickle
import pickle
with open('save/clf.pickle', 'wb') as f:    # 保存模型
    pickle.dump(clf, f)
```

```python
with open('save/clf.pickle', 'rb') as f:    # 读入模型
    clf2 = pickle.load(f)
    print(clf2.predict(X[:1]))
```

```python
# 方法二：joblib
import joblib
joblib.dump(clf, 'save/clf.pkl')
```
